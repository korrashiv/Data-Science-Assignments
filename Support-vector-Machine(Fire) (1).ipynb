```python
# applying standard scaler
from sklearn.preprocessing import StandardScaler
SC=StandardScaler()
X1[["FFMC"]]=SC.fit_transform(X1[["FFMC"]])
X1[["DMC"]]=SC.fit_transform(X1[["DMC"]])
X1[["DC"]]=SC.fit_transform(X1[["DC"]])
X1[["ISI"]]=SC.fit_transform(X1[["ISI"]])
X1[["temp"]]=SC.fit_transform(X1[["temp"]])
X1[["RH"]]=SC.fit_transform(X1[["RH"]])
X1[["wind"]]=SC.fit_transform(X1[["wind"]])
X1[["rain"]]=SC.fit_transform(X1[["rain"]])
X1[["area"]]=SC.fit_transform(X1[["area"]])
```

```python
# Splitting Train and Test
from sklearn.model_selection._split import train_test_split
X1_train,X1_test,y_train,y_test=train_test_split(X1,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear', C=1.0)
clf.fit(X1_train, y_train)
y_pred_train = clf.predict(X1_train)
y_pred_test  = clf.predict(X1_test)
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, y_pred_test)

print("Training Accuracy :",metrics.accuracy_score(y_train, y_pred_train).round(2))
print("Testing Accuracy:",metrics.accuracy_score(y_test, y_pred_test).round(2))

```

```python
# polynomial


from sklearn.svm import SVC
clf = SVC(kernel='poly', degree=1)
clf.fit(X1_train, y_train)
y_pred_train = clf.predict(X1_train)
y_pred_test  = clf.predict(X1_test)
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, y_pred_test)

print("Training Accuracy :",metrics.accuracy_score(y_train, y_pred_train).round(2))
print("Testing Accuracy:",metrics.accuracy_score(y_test, y_pred_test).round(2))

```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVC
from sklearn.decomposition import PCA

# Assuming X1_train and y_train are your training data with 11 features

# Fit an SVM model
clf = SVC(kernel='linear', C=1.0)
clf.fit(X1_train, y_train)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X1_train)

# Fit SVM on the reduced dimensions
clf.fit(X_pca, y_train)

```

```python
# Meshgrid for decision boundary plot
h = .02  # step size in the mesh
x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Plot decision boundary and data points
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_train, cmap=plt.cm.Paired, edgecolors='k')
```

```python
# Plot decision boundary and data points
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_train, cmap=plt.cm.Paired, edgecolors='k')

# Plot the direction of the best-fit line (in the original space)
coeff = clf.coef_[0]
line = pca.inverse_transform([coeff])  # Convert the direction back to the original space
plt.arrow(0, 0, line[0, 0], line[0, 1], color='red', linewidth=2.5, head_width=0.1)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('SVM Decision Boundary with Best-Fit Line (PCA)')
plt.show()
```
